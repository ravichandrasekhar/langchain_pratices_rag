```python
from langchain.text_splitter import CharacterTextSplitter

# Make a simple text of 150 characters
text = "abcdefghij" * 15   # "abcdefghij" is 10 chars → 15 * 10 = 150 chars

# Split into chunks of 100 characters, no overlap
splitter = CharacterTextSplitter(chunk_size=100, chunk_overlap=0)
chunks = splitter.split_text(text)

print("Total characters:", len(text))
print("Chunks produced:", len(chunks))
for i, c in enumerate(chunks, 1):
    print(f"Chunk {i}: length={len(c)} | content={c}")
```

```python
# 1) Default behavior (no explicit separator) -- splits by character count (default separator is a space)
char_splitter_default = CharacterTextSplitter(chunk_size=5, chunk_overlap=2)
chunks_default = char_splitter_default.split_text(text)
```

```python
chunks_default
```

```python
# 2) Explicit paragraph-first using separator="\n\n"
char_splitter_paragraphs = CharacterTextSplitter(separator="\n\n", chunk_size=100, chunk_overlap=20)
chunks_paragraphs = char_splitter_paragraphs.split_text(text)
```

```python
# Print results nicely
def print_chunks(title, chunks):
    print(f"\n--- {title} (count={len(chunks)}) ---")
    for i, c in enumerate(chunks, 1):
        print(f"[{i}] ({len(c)} chars) -> {repr(c[:120])}{'...' if len(c)>120 else ''}")

print_chunks("CharacterTextSplitter (default separator)", chunks_default)
print_chunks('CharacterTextSplitter (separator="\\n\\n")', chunks_paragraphs)
```
